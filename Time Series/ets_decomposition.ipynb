```python
import pandas as pd
from datetime import datetime
%matplotlib inline
```

```python
# ETS  - error, trend, seasonalityt
df = pd.read_csv('../../Time-series-analysis/Data/airline_passengers.csv',index_col='Month',parse_dates=True)
```

```python
df.head()
```

```python
# for ETS decomposition we should not have any NA
df.dropna()
```

```python
df.plot()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(df['Thousands of Passengers'], model='Multiplicative')
```

```python
result.plot();
```

```python
result.trend
```

```python
result.trend.plot(figsize=(12,6))
```
