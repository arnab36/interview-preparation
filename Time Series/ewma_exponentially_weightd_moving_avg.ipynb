```python
import pandas as pd
from datetime import datetime
%matplotlib inline
```

```python
df = pd.read_csv('../../Time-series-analysis/Data/airline_passengers.csv',index_col='Month')
```

```python
df.head()
```

```python
# See the index is  String index not a datetime index.
df.index
```

```python
df.dropna(inplace=True)
```

```python
df.index = pd.to_datetime(df.index)
```

```python
# Now you can see that the index is  datetime index
df.index
```

```python
# Remember datetime index will always start with first day of the month
df.head()
```

```python
list(df)
```

```python
# Calculating 6 month's simple moving average
df['6-month-SMA'] = df['Thousands of Passengers'].rolling(window=6).mean()
```

```python
df.plot()
```

```python
# Calculating 12 month's simple moving average
df['12-month-SMA'] = df['Thousands of Passengers'].rolling(window=12).mean()
```

```python
df.plot()
```

```python
# Exponentially Weighted Moving Avg
df['EWMA-12'] = df['Thousands of Passengers'].ewm(span=12).mean()
```

```python
df.plot()
```

```python
df[['Thousands of Passengers','EWMA-12']].plot()
```
