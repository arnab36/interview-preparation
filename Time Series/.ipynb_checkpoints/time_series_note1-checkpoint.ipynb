```python
df = pd.read_csv('../../Time-series-analysis/Data/starbucks.csv',index_col='Date',parse_dates=True)
```

```python
df.head()
```

```python
df.index
```

```python
# Rule A: daily -> yearly
df.resample(rule='A').mean()
```

```python
df.resample(rule='A').std()
```

```python
df.resample(rule='A').mean().plot.bar()
```

```python
title='Max closing price per month'
df['Close'].resample('M').max().plot.bar(figsize=(16,6), title=title)
```

```python
title='Max volume per month'
df['Volume'].resample('M').max().plot.bar(figsize=(16,6), title=title)
```

```python
df['Close'].plot()
```

```python
# Rolling average
df.rolling(window=7).mean()
```

```python
# Window remains the same
df.rolling(window=7).mean()['Close'].plot()
```

```python
df['Close'].plot()
df.rolling(window=7).mean()['Close'].plot()
```

```python
# General trend of the time series. If you decreese the window, then it will coincide with actual data
df['Close'].plot()
df.rolling(window=30).mean()['Close'].plot()
```

```python
# Window increases 
df['Close'].expanding().mean().plot()
```
