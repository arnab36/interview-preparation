```python
# Poisson Distribution
from scipy.stats import poisson
```

```python
# On a booking counter on the average 3.6 people come every 10 minutes on weekends. What is the probability of getting 7 
# people in 10 minutes?
```

```python
prob_of_seven = poisson.pmf(7, 3.6)
print(prob_of_seven)
```

```python
# Cumulative distribution
prob_seven_or_less = poisson.cdf(7, 3.6)
print(prob_seven_or_less)
```

```python
prob_more_than_seven = 1- prob_seven_or_less
print(prob_more_than_seven)
```

```python
# Let us plot the distribution
x_ax = np.arange(0,15)
y_ax = poisson.pmf(x_ax,3.6)
sns.barplot(x=x_ax,y=y_ax)
```

```python
# Let us plot the cumulative distribution
x_ax = np.arange(0,15)
y_ax = poisson.cdf(x_ax,3.6)
sns.barplot(x=x_ax,y=y_ax)
```

```python
# Normal DIstribution
from scipy.stats import norm
```

```python
# Perfume bottles are filled with the avg volume of 150 cc and the S.D of 2 cc. What % of bottles will have volume greater 
# than 153 cc?
```

```python
# norm.cdf(X, mean, SD)
prob_less_than_153_cc = norm.cdf(153, 150, 2)
print(prob_less_than_153_cc)
```

```python
prob_more_than_153_cc = 1 - prob_less_than_153_cc
print(prob_more_than_153_cc)
```

```python
# Perfume bottles are filled with the avg volume of 150 cc and the S.D of 2 cc. What % of bottles will have volume between  
# 148 cc and 152 cc?
```

```python
# As sigma = 2, we can see that we need arear between 1 Sigma left to the mean and right to the mean
prob_area_more_than_1_sigma_right_to_the_mean = 1 - norm.cdf(152, 150, 2)
print(prob_area_more_than_1_sigma_right_to_the_mean)
```

```python
prob_area_less_than_1_sigma_left_to_the_mean = norm.cdf(148, 150, 2)
print(prob_area_less_than_1_sigma_left_to_the_mean)
```

```python
area_between_1_sigma = prob_area_less_than_1_sigma_left_to_the_mean - prob_area_more_than_1_sigma_right_to_the_mean
print(area_between_1_sigma)
```
