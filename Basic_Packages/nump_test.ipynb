```python
# Binomial distribution
one_coin = np.random.binomial(1,0.5, 1000)
sns.countplot(x= one_coin)
```

```python
# Binomial distribution - 2 coins
two_coins = np.random.binomial(2,0.5, 1000)
```

```python
sns.countplot(x= two_coins)
```

```python
# Binomial distribution - 3 coins - Heads-> 0,1,2,3
three_coins = np.random.binomial(3,0.5, 1000)
sns.countplot(x= three_coins)
```

```python
# Poisson distribution (mean, samples)
np.random.poisson(3.6, 10)
```

```python
# let us plot
x = np.random.poisson(3.6, 1000)
sns.countplot(x=x)
```

```python
# Normal Distribution - (mean,SD)
np.random.normal(100,2)
```

```python
# (mean,SD, samples)
np.random.normal(100,2,10)
```

```python
# Let us plot it
sample_size = 1000000
x = np.random.normal(100,2,sample_size)
sns.histplot(x=x)
```

```python
# Sigma = 2
area_between_1_sigma_on_both_sides = sum((x > 98) & (x < 102))/sample_size
print(area_between_1_sigma_on_both_sides)
```
